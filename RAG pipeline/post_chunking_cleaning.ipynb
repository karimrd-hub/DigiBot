## Chunking

```python
# This code only lowercases the text

import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt')

def lowercase_exact_structure(input_file, output_file):
    """
    Lowercase text while preserving EXACT paragraph and line structure
    """
    with open(input_file, 'r', encoding='utf-8') as infile:
        content = infile.read()
    
    # Split content while preserving empty lines and exact structure
    lines = content.split('\n')
    processed_lines = []
    
    for line in lines:
        if not line.strip():  # Empty line (paragraph separator)
            processed_lines.append('')  # Keep exactly as empty line
            continue
        
        # Process non-empty lines while preserving original structure
        sentences = sent_tokenize(line)
        
        # Lowercase each sentence but preserve original spacing/formatting
        # by reconstructing the line with lowercase content
        processed_sentences = []
        current_pos = 0
        
        for sentence in sentences:
            # Find the original sentence in the line
            sentence_start = line.find(sentence, current_pos)
            if sentence_start == -1:
                # Fallback: just lowercase the sentence
                processed_sentences.append(sentence.lower())
                continue
            
            # Preserve any leading whitespace before the sentence
            leading_whitespace = line[current_pos:sentence_start]
            
            # Add the lowercase sentence with preserved whitespace
            processed_sentences.append(leading_whitespace + sentence.lower())
            current_pos = sentence_start + len(sentence)
        
        # Reconstruct the line exactly
        processed_line = ''.join(processed_sentences)
        # Add any trailing content that wasn't captured as sentences
        if current_pos < len(line):
            processed_line += line[current_pos:].lower()
        
        processed_lines.append(processed_line)
    
    # Write back with EXACT same line structure
    with open(output_file, 'w', encoding='utf-8') as outfile:
        outfile.write('\n'.join(processed_lines))

# Usage
input_filename = '../Scraping_Digico_Website/scraped_data/cleaned_text_content.txt'
output_filename = './data/cleaned_data.txt'

lowercase_exact_structure(input_filename, output_filename)
print(f"Processing complete with exact structure preservation.")
```
