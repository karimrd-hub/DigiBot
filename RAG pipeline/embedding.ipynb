```python
import os
import vertexai
from dotenv import load_dotenv
from vertexai.language_models import TextEmbeddingModel

# Load environment variables
load_dotenv()

# Get Vertex AI configuration from environment
PROJECT_ID = os.environ.get("GOOGLE_CLOUD_PROJECT")
LOCATION = os.environ.get("GOOGLE_CLOUD_LOCATION")

print(f"Using PROJECT_ID: {PROJECT_ID}")
print(f"Using LOCATION: {LOCATION}")

# Initialize Vertex AI
try:
    vertexai.init(project=PROJECT_ID, location=LOCATION)
    print("Vertex AI initialization successful")
except Exception as e:
    print(f"Failed to initialize Vertex AI: {str(e)}")

# Test input text
test_text = "Hello world! This is a test for Vertex AI embeddings."

# Initialize the embedding model
embedding_model = TextEmbeddingModel.from_pretrained(
    "publishers/google/models/text-embedding-005"
)

# Generate embedding
embedding_vector = embedding_model.get_embeddings([test_text])

# Print results
print("Embedding vector length:", len(embedding_vector[0].values))
print("First 10 values of the embedding:", embedding_vector[0].values[:10])
```
